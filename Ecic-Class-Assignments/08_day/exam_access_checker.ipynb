## Scenario

### User enters:

- item name
- quantity
- delivery time (day, night)

### Rules

- Some items unavailable at night
- Quantity limit check
- Normalize item name using .lower()

### Topics used

- nested if/else
- string modification
- and / or
- f-string output

```python
item = input("Enter your item name: ").strip().lower()
quantity = int(input("Enter your item quantity: "))
time = input("Enter delivery time in day/night: ").strip().lower()

if quantity > 0 and quantity <= 15:
    if time == "day":
        if item == "burger" or item == "pizza" or item == "fries":
            print(f"Sorry, {item} is not available at day.")
        else:
            print(f"item will be delivered at selected time")
    elif time == "night":
        print(f"Your order of {quantity} {item}(s) will be delivered at your selected time")
    else:
        print("Invalid delivery time entered.")
else:
    print("Quantity should be between 1 and 5.")
```
