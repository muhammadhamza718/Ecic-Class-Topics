### Q1: Calculate Area of Circle
Formula: Area = π × r²

Where:
- π (pi) = 3.14159
- r = radius of the circle

```python
# Area of Circle
pi = 3.14159
radius = 5

area = pi * radius ** 2
print("Area of Circle:", area)
```

### Q2: Calculate Average of 3 Numbers
Formula: Average = (num1 + num2 + num3) / 3

```python
# Average of 3 numbers
num1 = 10
num2 = 20
num3 = 30

average = (num1 + num2 + num3) / 3
print("Average:", average)
```

### Q3: Calculate Simple Interest
Formula: SI = (P × R × T) / 100

Where:
- P = Principal amount
- R = Rate of interest per year
- T = Time period in years

```python
# Simple Interest
P = 1000  # Principal
R = 5     # Rate
T = 2     # Time in years

si = P * R * T / 100
print("Simple Interest:", si)
```
