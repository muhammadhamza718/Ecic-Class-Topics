```python
Marks=int(input("Enter your numbers= "))

if Marks <= 40:
    print("Your grade is F")
elif Marks <= 50:
    print("Your grade is E")
elif Marks <= 60:
    print("Your grade is C")
elif Marks <= 70:
    print("Your grade is B")
elif Marks <= 80:
    print("Your grade is A")
elif Marks <= 100:
    print("Your grade is A+")
else:
    print("unknown number")
    


```
