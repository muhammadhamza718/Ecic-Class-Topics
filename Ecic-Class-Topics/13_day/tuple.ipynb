## Tuple

#### - A tuple is an ordered, immutable collection of element in python

## Examples: 

### 01: Simple tuples

```python
t = (1, 2, 3)
```

### 02: Empty tuple

```python
t = ()
```

### 03: Single element tuple

```python
t = (3)
```

### 04: Multiple element tuples

```python
t = (1, "Python", 3.5)
```

### 05: Without brackets tuples 

```python
t = 1, 2, 3
```

### 06: indexing tuple

```python
t (10, 20, 30)
print(t[0])
```

### 07: Negative indexing tuple

```python
print(t[-1])
```

### 08: Slicing with tuple index

```python
print(t[0:2])
```

```python
name = ["Hamza", "Ubaid", "khuram"]
print(name)
name.append("Ubaid")
print(name)

name_in_set = set(name)
print(name_in_set)
```

```python
name = ["Hamza", "Ubaid", "khuram"]
print(name)
name.append("Ubaid")
print(name)

name_in_set = set(name)
print(name_in_set)

print(len(name_in_set))
```
