## Combining Multiple Operators

### You can combine multiple logical operators in a single expression python evaluates not first then and or 
### Use parenthesis for clarity 

#### And:
- T - F = F
- T - T = T
- F - F = F

#### Or:
- T - F = T
- T - T = T
- F - F = F

### Not:
- T -> F
- F -> T

### Example:

```python
temperature = 25
raining = False
weekend = True

if(temperature > 20 and not raining) or weekend:
    print("Its a great day for outdoors activities")
else:
    print("Not good for outdoor activities")
```

### Example 02:

```python
user_active = True
if not user_active:
    print("User is not active")
else:
    print("User is active")
```

### Example 03:

```python
age = 24
has_promo_code = True

if age < 18 or age > 65 or has_promo_code:
    print("Discount Applies")
else:
    print("Discount Not Applies")
```

### Example 04:

```python
Student = True
Scholarship = True

if Student and not Scholarship:
    print("Fee Discount Applies")
else:
    print("Standard Fee Applies")
```
