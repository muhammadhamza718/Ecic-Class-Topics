## Sets:

- A sets is unodered collection of unique elements
- sets are mutable
- sets cannot store the duplicated values
- you can add or remove variable but do not allow duplicate variable

## Sets Operators:

- I -> Union
- & -> intersection
- \- -> difference
- ^ -> symetic difference

1. The union() method returns a new set with all items from both sets.
2. The intersection() method returns a set that contains the similarity between two or more sets.
3. The difference() method returns a set that contains the difference between two sets.
4. The symmetric_difference() method returns a set that contains all items from both set, but not the items that are present in both sets.

```python
A = {1, 2, 3, 4}
b = {2, 4, 6, 8}

C = A.union(b)
print(C)
```

```python
A = {1, 2, 3, 4}
b = {2, 4, 6, 8}

C = A.intersection(b)
print(C)
```

```python
A = {1, 2, 3, 4}
b = {2, 4, 6, 8}

C = A.difference(b)
print(C)
```

```python
A = {1, 2, 3, 4}
b = {2, 4, 6, 8}

C = A.symmetric_difference(b)
print(C)
```
