## Casing Methods In Python

1. capitalize()
2. Casefold()
3. Center()

### Capitalize:

- Capitalize means it can change the string first character into capital letter and all the remaining into lowercase if they are capital

```python
text = "hello, and Welcome to my work"
x = text.capitalize()
print(x)
```

```python
text = "hello, and Welcome to my work"
x = text.casefold()
print(x)
```

```python
text = "hello"
x = text.center(20, " ")
print(x)
```
