## Python Logical Operators

### if else statement using and. or and not operators

```python
pedestriam_signals = "Flashing"
if pedestriam_signals == "walk" and pedestriam_signals == "Flashing":
    print("Cross the road")
else:
    print("Do not cross the road")
```

```python
pedestriam_signals = "Flashing"
if pedestriam_signals == "walk" or pedestriam_signals == "Flashing":
    print("Cross the road")
else:
    print("Do not cross the road")
```

```python
pedestriam_signals = "Flashing"
if pedestriam_signals == "walk" or pedestriam_signals == "Flashing":
    print("Cross the road")
else:
    print("Do not cross the road")
```

```python
pedestriam_signals = "Flashing"
active = "Not flashing"
if pedestriam_signals is not active:
    print("Cross the road")
else:
    print("Do not cross the road")
```

```python
pedestriam_signals = "Flashing"
active = "Flashing"
if pedestriam_signals is not active:
    print("Cross the road")
else:
    print("Do not cross the road")
```

```python
a = "Flashing"
b = a

if a is b:
    print("True")
else:
    print("False")
```
