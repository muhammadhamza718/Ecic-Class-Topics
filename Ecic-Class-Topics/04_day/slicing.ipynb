# String Slicing Practice

Slicing allows you to extract a "chunk" of characters from a string.
The format is `[start : stop : step]`.

**Remember:**
*   The `start` index is **included**.
*   The `stop` index is **excluded** (it stops *before* this index).
*   If you leave `start` empty, it starts from the beginning.
*   If you leave `stop` empty, it goes to the end.

---

### Exercise 1: Basic Slice
Create a variable called `text` containing `"PythonProgramming"`.
Print only the word `"Python"` by slicing from the beginning to the correct index.

```python
text = "Python Programming"
print(text[0:6])
```

### Exercise 2: Slicing from a Start Point
Using the same `text` variable (`"PythonProgramming"`), print the word `"Programming"`.
(Hint: Start at the letter 'P' of Programming and go to the end).

```python
text = "Python Programming"
print(text[7:])
```

### Exercise 3: Slicing a Middle Section
Create a variable called `sentence` containing `"I love coding"`.
Print only the word `"love"` using slicing.

```python
sentence = "I love coding"
print(sentence[2:6])
```

### Exercise 4: Negative Slicing
Create a variable called `world` containing `"Hello World"`.
Print the word `"World"` using **negative indices** (counting from the end).

```python
world = "Hello World"
print(world[-5:])
```

### Exercise 5: Step Slicing
Create a variable called `numbers` containing `"0123456789"`.
Print every **second** number (`"02468"`) by using a step of 2.

```python
numbers = "0123456789"
print(numbers[::2])
```

### Exercise 6: Reversing a String
Create a variable called `word` containing `"stressed"`.
Print the string in reverse to see what it spells!

```python
word = "stressed"
print(word[::-1])
```
