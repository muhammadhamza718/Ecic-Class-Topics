# If, Elif, Else & Comparison Operators

In these exercises, you will practice using **comparison operators** together with **if**, **elif**, and **else** statements.

**Remember:**
*   `if`: Tests the first condition.
*   `elif`: Tests a new condition if the previous ones were False.
*   `else`: Runs if ALL previous conditions were False.

---

### Exercise 1: Equal To (`==`)
1. Create a variable `password` with the value `"1234"`.
2. Write an **if/else** statement:
   *   If `password` **equals** `"1234"`, print `"Access Granted"`.
   *   **Else**, print `"Access Denied"`.

```python
password = "1234"
if password == "1234":
    print("Access Granted")
else:
    print("Access Denied")
```

### Exercise 2: Not Equal To (`!=`)
1. Create a variable `color` with the value `"Red"`.
2. Write an **if/else** statement:
   *   If `color` does **not equal** `"Blue"`, print `"This is not Blue"`.
   *   **Else**, print `"This is Blue"`.

```python
color = "Red"
if color != "Blue":
    print("This is not Blue")
else:
    print("This is Blue")
```

### Exercise 3: Greater Than (`>`)
1. Create a variable `score` with the value `100`.
2. Write an **if/else** statement:
   *   If `score` is **greater than** `50`, print `"Passing Score"`.
   *   **Else**, print `"Failing Score"`.

```python
score = 100
if score > 50:
    print("Passing Score")
else:
    print("Failing Score")
```

### Exercise 4: Less Than (`<`)
1. Create a variable `temperature` with the value `-5`.
2. Write an **if/else** statement:
   *   If `temperature` is **less than** `0`, print `"It is freezing"`.
   *   **Else**, print `"It is NOT freezing"`.

```python
temperature = -5
if temperature < 0:
    print("It is freezing")
else:
    "It is NOT freezing"
```

### Exercise 5: Greater Than or Equal To (`>=`)
1. Create a variable `age` with the value `18`.
2. Write an **if/else** statement:
   *   If `age` is **greater than or equal to** `18`, print `"Adult"`.
   *   **Else**, print `"Minor"`.

```python
age = 18
if age >= 18:
    print("Adult")
else:
    print("Minor")
```

### Exercise 6: Less Than or Equal To (`<=`)
1. Create a variable `price` with the value `20`.
2. Write an **if/else** statement:
   *   If `price` is **less than or equal to** `20`, print `"I can buy this"`.
   *   **Else**, print `"Too expensive"`.

```python
price = 20
if price <= 20:
    print("I can buy this")
else:
    print("Too expensive")
```

### Exercise 7: The Challenge (If, Elif, Else)
1. Create a variable `number` with **any** value you want.

   *   If `number` is **greater than** `0`: Print `"Positive"`.
   *   **Elif** `number` is **less than** `0`: Print `"Negative"`.
