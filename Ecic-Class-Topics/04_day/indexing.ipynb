# String Indexing Practice

Practice using variables, `print()`, and indexing to solve these problems. 
Remember: **Python is case-sensitive** (commands like `print` must be lowercase).

---

### Exercise 1: First Character
Create a variable called `word` containing `"Python"`.
Print the **first** character.

```python
word = "Python"
print(word[0])
```

### Exercise 2: Last Character (Negative Indexing)
Create a variable called `message` containing `"Hello"`.
Print the **last** character using a **negative index**.

```python
message = "Hello"
print(message[-1])
```

### Exercise 3: Accessing a Specific Character
Create a variable called `fruit` containing `"Orange"`.
Print the character `a` from this string by using the correct index number.

```python
fruit = "Orange"
print(fruit[2])
```

### Exercise 4: Slicing
Create a variable called `text` containing `"Wonderful"`.
Use slicing to print only the first part: `"Wonder"`.

```python
text = "Wonderful"
print(text[0:6])
```

### Exercise 5: Finding Numbers
Create a variable called `numbers` containing the string `"0123456789"`.
Print the digit `5` using the correct index.

```python
numbers = "0123456789"
print(numbers[5])
```
