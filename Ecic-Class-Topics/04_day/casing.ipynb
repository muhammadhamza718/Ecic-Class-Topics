# String Casing Exercises (Solutions)

Here are the methods to change the "case" of a string:
*   `.upper()`: Converts to ALL CAPS.
*   `.lower()`: Converts to all lowercase.
*   `.title()`: Capitalizes the First Letter of Each Word.

**Run the cells below to see how they work!**

### Exercise 1: Uppercase
Convert the string `"python"` to uppercase.

```python
text = "python"
print(text.upper())
```

### Exercise 2: Lowercase
Convert the string `"PYTHON"` to lowercase.

```python
text = "PYTHON"
print(text.lower())
```

### Exercise 3: Title Case
Convert the string `"python programming"` to title case.

```python
text = "python programming"
print(text.title())
```
