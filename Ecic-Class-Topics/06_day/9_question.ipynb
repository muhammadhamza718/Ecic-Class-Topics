```python
weight = int(input("Enter your weight in kilogram: "))
height = int(input("Enter your height in centimeter: "))

bmi = height * 2 / weight 
print(bmi)

if bmi < 18.5:
    print("You are Underweight")
else:
    if bmi < 25:
        print("You are Normal")
    else:
        if bmi < 30:
            print("You are Overweight")
        else:
            print(" You are Overweight")
```
