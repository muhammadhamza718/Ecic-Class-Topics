## Python Modify String

### Use: Data Cleaning
- trimming -> Removing accidental spaces
- normalizing -> Converting to lowercase
- Removing noise -> Deleting special characters

### Example 01:

```python
a = "Hello, World!"
print(a.upper())
```

### Example 02:

```python
a = "  Hello, World!  "
print(a.strip())
```

### Example 03:

```python
a = "Hello, World!"
print(a.replace("H", "J"))
```

### Example 04:

```python
a = "Hello, World!"
print(a.split(","))
```
